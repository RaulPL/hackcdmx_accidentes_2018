```python
import ujson
import joblib
import hdbscan
import cartopy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
from pathlib import Path
import geoviews as gv
import holoviews as hv

from data_utils import load_axa_data

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)

%matplotlib inline
hv.extension('bokeh')
```

```python
# Load datos axa
df_axa = pd.read_parquet('../data/processed/accidentes_axa.parquet', engine='pyarrow')
print(df_axa.shape)
df_axa.head()
```

# Perfil de accidentes

```python
with plt.style.context('seaborn-talk'):
    df_aux = df_axa[['dia_semana', 'hora', 'siniestro_n_10_0']].groupby(['dia_semana', 'hora'], as_index=False).siniestro_n_10_0.count()
    df_aux = df_aux.pivot(index='dia_semana', values='siniestro_n_10_0', columns='hora')
    df_aux = df_aux.reindex(['LUNES', 'MARTES', 'MIERCOLES', 'JUEVES', 'VIERNES', 'SABADO', 'DOMINGO'])
    fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(18, 6))
    sns.heatmap(df_aux, cmap='viridis', annot=True, fmt='d', ax=ax)
    plt.tight_layout()
```

# Serie de tiempo de accidentes

```python

df_incidentes = df_axa.groupby('fecha', as_index=False).siniestro_n_10_0.count()
df_incidentes = df_incidentes.rename(columns={'siniestro_n_10_0': 'incidentes'})

chart_line = alt.Chart(df_incidentes).mark_line().encode(
    x='fecha:T',
    y='incidentes:Q'
).interactive()

chart_point = alt.Chart(df_incidentes).mark_point(size=10).encode(
    x='fecha:T',
    y='incidentes:Q'
).interactive()

chart = chart_line + chart_point
chart.configure_view(
    height=500,
    width=1280
)
```
